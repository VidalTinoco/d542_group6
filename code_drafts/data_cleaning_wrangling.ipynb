### Find mode conversation language and filter out non-English conversations

```python
# want to see if prompts contain code
n = 80

rows = conversations['Prompt'].iloc[n:n+10]
for i, row in enumerate(rows):
    print(f"Row: {n + i}")
    print(row)
```

```python
# noticed that there's null prompts/answers
null_prompts = conversations[conversations['Prompt'].isnull()]
null_prompts
```

```python
# check if rows with null prompts have null answers
null_prompts_with_null_answers = null_prompts[null_prompts['Answer'].isnull()] # check if 'Answer' column is also null in these rows

# compare the lengths of the two DataFrames
all_null = len(null_prompts) == len(null_prompts_with_null_answers)
all_null
```

```python
# drop rows where 'Prompt' and 'Answer' is null
conversations = conversations[~(conversations['Prompt'].isnull() & conversations['Answer'].isnull())]
conversations.info()
```

```python
conversations['conversation_id'].value_counts() # drops from 373
```

```python
from langdetect import detect, LangDetectException

def detect_language(prompt):
    try:
        # Detect the language of the input prompt
        language = detect(prompt)
        return language
    except LangDetectException as e:
        print(f"Language detection error: {e}")
        return None  # Return None if language detection fails

# apply the detect_language function to the 'Prompt' column and create a new 'Detected_Language' column
conversations['Detected_Language'] = conversations['Prompt'].apply(detect_language)

# check the updated DataFrame
conversations.head()
```

```python
# checking rows where no language was detected
undetected = conversations[conversations['Detected_Language'].isnull()]
undetected.head()
```

```python
# group by 'conversation_id' and find the mode of 'Detected_Language' for each conversation
mode_languages = conversations.groupby('conversation_id')['Detected_Language'].agg(lambda x: x.mode()[0] if not x.mode().empty else None)

# get a list of conversation_ids where the mode language is 'en'
english_conversations = mode_languages[mode_languages == 'en'].index.tolist()

# number of english conversations (total is 369)
print(len(english_conversations))
```

```python
# drop non-English conversations
conversations = conversations[conversations['conversation_id'].isin(english_conversations)]
conversations.head()
```
