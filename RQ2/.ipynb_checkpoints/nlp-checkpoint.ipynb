### Research Question: Can we identify patterns in the prompts developers use when interacting with ChatGPT, and how do these patterns correlate with the success of issue resolution?

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/data_clean/issuedata_cleaned.csv')
df.info()
```

```python
df['conversation_id'].value_counts()
```

```python
df.head()
```

```python
conversations = df.copy()
conversations = conversations[['conversation_id', 'State', 'numPrompts']].drop_duplicates()

for value, count in conversations['numPrompts'].value_counts().items():
    print(f"Value: {value}, Count: {count}")
```

```python
chart = alt.Chart(conversations).mark_bar().encode(
    x=alt.X('numPrompts:O', title='Number of Prompts'),  # Use numPrompts as the x-axis
    y=alt.Y('count():Q', title='Number of Conversations'),  # Count number of conversations
    color='State'  # Color by Issue_Status (open/closed)
).properties(
    title="Issue Status vs Number of Prompts"
)

chart.show()
```

```python
import altair as alt

conversations = df.copy()
conversations = conversations[['conversation_id', 'State', 'numPrompts']].drop_duplicates()

# Create box plot
chart = alt.Chart(conversations).mark_boxplot().encode(
    x=alt.X('State:N', title='Issue Status'),
    y=alt.Y('numPrompts:Q', title='Number of Prompts'),
    color='State',  # Color by Issue_Status (open/closed)
    tooltip=['State', 'numPrompts']  # Add tooltips for better interactivity
).properties(
    title="Issue Status vs Distribution of Number of Prompts"
)

chart.show()
```
