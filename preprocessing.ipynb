```python
import pandas as pd
from pandas import json_normalize
import numpy as np
import base64
```

```python
file_sharings = 'data/snapshot_20230831/20230831_072722_file_sharings.json'
commit_sharings = 'data/snapshot_20230831/20230831_063412_commit_sharings.json'
issue_sharings = 'data/snapshot_20230831/20230831_061759_issue_sharings.json'
pr_sharings = 'data/snapshot_20230831/20230831_060603_pr_sharings.json'
discussion_sharings = 'data/snapshot_20230831/20230831_061926_discussion_sharings.json'
hn_sharings = 'data/snapshot_20230831/20230831_073827_hn_sharings.json'
chatgpt_link_sharings = 'data/snapshot_20230831/ChatGPT_Link_Sharing.csv'
```

### Pick file and read into dataframe

```python
file = issue_sharings

df = pd.read_json(file)
df = json_normalize(df['Sources']) # unpack 'Sources' json
df.info() # note number of entries
```

```python
df.head()
```

### Unpack nested jsons into dataframes

```python
# explode list data in 'ChatgptSharing' column and normalize json to dataframe
gpt = json_normalize(df['ChatgptSharing'].explode())

# add a conversation_id column to the gpt DataFrame
gpt['conversation_id'] = gpt['Title'] + '_' + gpt['Mention.MentionedURL'] # 'Mention.MentionedURL' links to 'URL' in parent df

# replace NaN values in Conversations with empty lists
gpt['Conversations'] = gpt['Conversations'].apply(lambda x: x if isinstance(x, list) else [])

gpt.info() # note number of entries compared to parent df
```

```python
gpt.head()
```

```python
# check if number of unique values per column looks correct
gpt.describe(include='all')
```

```python
# explode list data in 'Conversations' column and normalize json to dataframe
conversations = json_normalize(gpt['Conversations'].explode())

# propagate 'conversation_id' from gpt to the conversations DataFrame
conversations['conversation_id'] = gpt['conversation_id'].repeat(gpt['Conversations'].apply(len)).reset_index(drop=True)

conversations.info() # note number of entries compared to parent df
```
